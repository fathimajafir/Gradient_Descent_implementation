```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X=np.arange(10)
Y=(X-5)**2
print(X,Y)
```

```python
#Given a function f(x), find the value of x which minimizes f
```

```python
#visualisation
plt.style.use("seaborn")
plt.plot(X,Y)
plt.xlabel('X')
plt.ylabel('F(X)')
plt.show()
```

```python
x=0
lr=0.1 #learning rate
error=[]
plt.plot(X,Y)
#50 steps in the downhill direction 
for i in range(50):
  grad=2*(x-5)
  x=x-lr*grad
  y=(x-5)**2
  error.append(y)
  plt.scatter(x,y)

  print(x)
```
